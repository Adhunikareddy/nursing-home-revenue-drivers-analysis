```python
import pandas as pd
import os

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
# columns = ['Net_Income', 'Gross_Revenue', 'Total_Costs', 'Less_Total_Operating_Expense']
# columns = ['Net_Income', 'Gross_Revenue', 'Total_Costs', 'Less_Total_Operating_Expense']


df_all_years = []

for year in years:
    print(f"Processing {year} Cost Report...")
    # Load cost report
    cost_file = f'Data/{year}_CostReport.csv'
    df_cost = pd.read_csv(cost_file, encoding='cp1252', low_memory=False)

    if year in [2020, 2021]:
        df_cost.columns = df_cost.columns.str.replace(' ', '_')
    
    df_cost = df_cost[['Provider_CCN'] + columns].copy()

    # Fill missing values with column median
    df_cost[columns] = df_cost[columns].fillna(df_cost[columns].median(numeric_only=True))

    # Remove outliers using IQR
    for col in columns:
        Q1 = df_cost[col].quantile(0.25)
        Q3 = df_cost[col].quantile(0.75)
        IQR = Q3 - Q1
        df_cost = df_cost[(df_cost[col] >= Q1 - 1.5 * IQR) & (df_cost[col] <= Q3 + 1.5 * IQR)]

    # Read penalty file for the same year
    penalty_file = f'Data/Penalties_{year}.csv'
    if os.path.exists(penalty_file):
        print(f"Processing {year} Penalty Data...")
        df_penalty = pd.read_csv(penalty_file, encoding='cp1252', low_memory=False)

        # Handle missing values
        # df_penalty['fine_amt'] = df_penalty['fine_amt'].fillna(df_penalty['fine_amt'].mean())
        df_penalty['fine_amt'] = df_penalty['fine_amt'].fillna(0)
        

        # Outlier capping using IQR
        Q1 = df_penalty['fine_amt'].quantile(0.25)
        Q3 = df_penalty['fine_amt'].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_penalty['fine_amt'] = df_penalty['fine_amt'].clip(lower=lower_bound, upper=upper_bound)

        # Group by provider and sum penalties
        df_penalty_grouped = df_penalty.groupby('provnum', as_index=False)['fine_amt'].mean()
        df_penalty_grouped.rename(columns={'provnum': 'Provider_CCN', 'fine_amt': 'total_fine_amt'}, inplace=True)

        # Ensure matching key format
        df_penalty_grouped['Provider_CCN'] = df_penalty_grouped['Provider_CCN'].astype(str)
        df_cost['Provider_CCN'] = df_cost['Provider_CCN'].astype(str)

        # Merge cost report with penalty data
        df_merged = pd.merge(df_cost, df_penalty_grouped, on='Provider_CCN', how='left')
        df_merged['total_fine_amt'] = df_merged['total_fine_amt'].fillna(0)  # fill providers with no penalties

        # Add year column
        df_merged['Year'] = year
        df_all_years.append(df_merged)

    else:
        print(f"Penalty file for {year} not found. Skipping merge for this year.")

# Combine all years into one DataFrame
df_combined_all = pd.concat(df_all_years, ignore_index=True)


summary = df_combined_all.groupby('Year')[
    ['Net_Income', 'Gross_Revenue', 'Total_Costs', 'Less_Total_Operating_Expense', 'total_fine_amt']
].sum().reset_index()

print(summary)
```

```python
import pandas as pd
import os

# Years to process
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

# Financial columns to retain
columns = ['Net_Income', 'Gross_Revenue']

# Output dictionary
merged_dataframes = {}

for year in years:
    print(f"Processing year {year}...")

    # Load cost report
    cost_path = f"Data/{year}_CostReport.csv"
    cost_df = pd.read_csv(cost_path,encoding='cp1252', low_memory=False)

    # Rename columns for 2020–2021 formatting consistency (optional here but harmless)
    cost_df.columns = cost_df.columns.str.replace(' ', '_')

    # Select available financial columns
    available_cols = ['Provider_CCN'] + [col for col in columns if col in cost_df.columns]
    cost_df = cost_df[available_cols].copy()

    # Fill missing numeric values with column median
    cost_df[cost_df.columns[1:]] = cost_df[cost_df.columns[1:]].fillna(
        cost_df[cost_df.columns[1:]].median(numeric_only=True)
    )

    # IQR-based outlier filtering
    for col in cost_df.columns[1:]:
        Q1 = cost_df[col].quantile(0.25)
        Q3 = cost_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        cost_df = cost_df[(cost_df[col] >= lower_bound) & (cost_df[col] <= upper_bound)]
    # Add standardized Provider ID and year
    cost_df['Provider_CCN'] = cost_df['Provider_CCN'].astype(str).str.zfill(6)
    cost_df['Year'] = year

    # Load provider info
    provider_path = f"Data/ProviderInfo_{year}.csv"
    provider_df = pd.read_csv(provider_path, encoding='cp1252')

    # Clean and prepare provider info
    provider_df.columns = provider_df.columns.str.strip()
    provider_df['provnum'] = provider_df['provnum'].astype(str).str.zfill(6)
    provider_df['Year'] = pd.to_datetime(provider_df['FILEDATE'], errors='coerce').dt.year
    provider_df = provider_df[provider_df['Year'] == year]

    # Merge on Provider ID and Year
    merged_df = pd.merge(
        cost_df,
        provider_df,
        left_on=['Provider_CCN', 'Year'],
        right_on=['provnum', 'Year'],
        how='inner'
    )

    # Store in dictionary
    merged_dataframes[year] = merged_df


    
# You now have:

# merged_dataframes[2015].columns

# for i in range(1, 7):
#     print(merged_dataframes[i+2014].columns)
```
